```python
pattern=r'@\w+ '
for i in range(len(data_test)):
    data_test['text'].iloc[i] = re.sub(pattern,'', data_test['text'].iloc[i], flags=re.MULTILINE)
data_test.head()
```

```python
import re
pattern=r'[^A-Za-z ]'
for i in range(len(data_test)):
    data_test['text'].iloc[i] = re.sub(pattern,'', data_test['text'].iloc[i], flags=re.MULTILINE)
print (data_test['text'][0])
data_test.head()
```

```python
import re
import string
import csv
```

```python
reader = csv.reader(open('Corpus/normalisasi.csv', 'r'))
d = {}
for row in reader:
    k,v= row
    d[str.lower(k)] = str.lower(v)
    #print d[k]
pat = re.compile(r"\b(%s)\b" % "|".join(d))
for i in range(len(data_test)):
    text = str.lower(data_test['text'].iloc[i])
    text = pat.sub(lambda m: d.get(m.group()), text)
    #print text
    data_test['text'].iloc[i]=text
print (data_test.text[0])
data_test.head(10)
```

```python
import nltk 
from nltk.corpus import stopwords
reader=pd.read_excel('Corpus/stopword_id.xls',header=None)
cachedStopWords = set(stopwords.words("english"))
cachedStopWords.update(reader[0][:])
for i in range(len(data_test)):
    sent=data_test['text'].iloc[i]
    kt=" ".join([word for word in sent.split() if word not in cachedStopWords])
    data_test['text'].iloc[i]=kt
print (data_test.text[0])
data_test.head(10)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
for i in range(len(data_test)):
    sent=data_test['text'].iloc[i]
    output = stemmer.stem(sent)
    data_test['text'].iloc[i]=output
print (data_test.text[0])
data_test
```
